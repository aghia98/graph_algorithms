```python
def DFS_iterative(graph, node): #explore all reachable nodes starting from node using DFS
    stack=[node]
    explored=[node]
    while len(stack)>0:
        node=stack.pop()
        print(node)
        for neighbour in graph[node]:
            if neighbour not in explored:
                stack.append(neighbour)
                explored.append(neighbour)
            
def DFS_recursive(graph, node): #explore all reachable nodes
    def DFS_recursive_call(graph, node, explored):
        print(node)
        for neighbour in graph[node]:
            if neighbour not in explored:
                DFS_recursive_call(graph, neighbour, explored)
                
    explored=[node]
    DFS_recursive_call(graph, node, explored)      
```

```python
def BFS(graph, node): #explore all reachable nodes starting from node
    stack=[node]
    explored=[node]
    while len(stack)>0:
        node=stack.pop(0)
        print(node)
        for neighbour in graph[node]:
            if neighbour not in explored:
                stack.append(neighbour)
```

```python
def has_path_iterative(graph, src, dst): #is there a path between src node and dst node
    stack=[src]
    explored=[src]
    while len(stack)>0:
        node=stack.pop()
        if node==dst: return True 
        for neighbour in graph[node]:
            if neighbour not in explored:
                if neighbour==dst: return True
                stack.append(neighbour)
                explored.append(neighbour)
    return False

def has_path_recursive(graph, src, dst):
    def has_path_recursive_call(graph, src, dst, explored):
        if src==dst: return True
        for neighbour in graph[src]:
            if neighbour not in explored:
                explored.append(neighbour)
                if has_path_recursive_call(graph, neighbour, dst, explored):
                    return True #break the for loop
        return False
    explored=[src]
    has_path_recursive_call(graph, src, dst, explored)  
```

```python
def n_connected_components(graph): #return the number of connected components in a graph
    node=list(graph.keys())[0]
    stack=[node]
    explored=[node]
    count=0
    remaining_nodes= set(graph.keys()).difference(set(explored))
    while remaining_nodes!=set({}):
        while len(stack)>0:
            node=stack.pop()
            for neighbour in graph[node]:
                if neighbour not in explored:
                    stack.append(neighbour)
                    explored.append(neighbour)
        count+=1
        remaining_nodes=set(graph.keys()).difference(set(explored))
        if remaining_nodes!=set({}):
            stack=[list(remaining_nodes)[0]]
            explored.append(list(remaining_nodes)[0])
    return count
```

```python
def largest_component(graph): #return a set nodes that is the largest component of the graph
    node=list(graph.keys())[0]
    largest_component=set({})
    stack=[node]
    explored=[node]
    component=set({node})
    remaining_nodes= set(graph.keys()).difference(set(explored))
    while remaining_nodes!=set({}):
        while len(stack)>0 :
            node=stack.pop()
            for neighbour in graph[node]:
                if neighbour not in explored:
                    stack.append(neighbour)
                    explored.append(neighbour)
                    component.add(neighbour)
        if len(component)>len(largest_component):
            largest_component=component.copy()
        remaining_nodes=set(graph.keys()).difference(set(explored))
        if remaining_nodes!=set({}):
            stack=[list(remaining_nodes)[0]]
            explored.append(list(remaining_nodes)[0])
            component=set({list(remaining_nodes)[0]})
    return largest_component
```

```python
def shortest_path(graph, src, dst): #return the length of the shortest path between src and dst
    stack=[(src,0)]
    explored=[src]
    while len(stack)>0 :
        node,distance=stack.pop(0) #couple (node,disatnce)
        if node==dst: return distance
        for neighbour in graph[node]:
            if neighbour not in explored:
                stack.append({neighbour,distance+1})
                explored.append(neighbour)
    return float('inf')
```
